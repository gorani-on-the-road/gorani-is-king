책에서는 SDGclassifier의 손실함수를 loss = 'log'(로지스틱 손실 함수,logistic loss function)을 사용했지만 원래 SDGclassifier의 기본 손실함수는 힌지손실(hinge loss)인점을 보아 힌지손실(hinge loss function)에 대해 조사해봄

※책에서는 자세하게 다루지않고 힌지 손실(hinge loss function)은 서포트 벡터 머신(suppoer vector machine,줄여서 SVM)라는 또 다른 머신러닝 알고리즘을 위한 손실함수다 까지밖에 나오지 않음<br><br>

힌지손실(hinge loss function)이란?


image.png

힌지 손실은 분류기, **특히 SVM을 훈련하는 데 사용되는 손실 함수이다.** 위의 그림은 그것이 어떻게 생겼는지에 대한 정말 좋은 시각화자료<br>

x 축은 단일 인스턴스의 경계로부터의 거리를 나타내고 y 축은 거리에 따라 함수가 발생하는 손실 크기 또는 패널티를 나타냄

<br>※유의사항<br>

a. x 축의 점선은 숫자 1을 나타냅니다. 즉, 경계에서 인스턴스의 거리가 1보다 크거나 같을 때 손실 크기는 0입니다.<Br>

b. 경계로부터의 거리가 0이면 (즉, 인스턴스가 문자 그대로 경계에 있음) 손실 크기는 1이됩니다.<br>

c. 올바르게 분류 된 포인트는 손실 크기가 작거나없는 반면 잘못 분류 된 인스턴스는 손실 크기가 높습니다.<br>

d. 경계에서 음의 거리는 높은 힌지 손실을 초래합니다. 이것은 본질적으로 우리가 경계의 잘못된쪽에 있으며 인스턴스가 잘못 분류된다는 것을 의미합니다.<br>

e. 플립 크기에서 경계에서 양의 거리는 힌지 손실이 적거나 힌지 손실이 전혀 발생하지 않으며 경계에서 멀어 질수록 (그리고 오른쪽에서) 힌지 손실이 낮아집니다<br>

image.png

이 그래프는 본질적으로 이전 시각화에서 관찰 한 내용을 강화합니다. 다시 한 번 인스턴스의 거리가 1보다 크거나 같으면 힌지 손실이 0임을 알 수 있습니다.

지점이 경계에있을 때 힌지 손실은 1 (녹색 상자로 표시됨)이고 경계로부터의 거리가 음수 (경계의 잘못된쪽에 있음을 의미)이면 점점 더 큰 힌지 손실을 얻습니다.
